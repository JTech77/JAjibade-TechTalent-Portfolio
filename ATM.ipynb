```python
#LogIn
def UserPin(pin):
        if pin == "1234":
            return True
        else:
            return False


def UserLogIn():
        pin = input("Please Enter Your 4 Digit Pin: ")
        if UserPin(pin):
            print("Welcome To The Bank of Joshua\n" "Press 1 for Current Balance\n", "Press 2 for Withdrawal\n", "Press 3 for Deposit\n", "Press 9 to Exit\n")
            return True
        else:
            print("Invalid Pin Number, Try Again")
            return False
UserLogIn()

#Variables For Menus

CurrentBalance = 5000
WD = 0
DP = 0
MM = input("Select 1, 2, 3 or 9")

#CurrentBalance
def Programme(MM):
    if MM =="1":
        print("You currently have", CurrentBalance)
        MM = input("Select 1, 2, 3 or 9")

#Withdrawls

    while MM =="2":
        print("How much do you want to withdraw?")
        WD = int(input("Enter Amount:"))
        if WD > CurrentBalance:
            print("Invalid, you do not have enough money.")
            MM = input("Select 1, 2, 3 or 9")
            continue
        else:
            print("You now have", CurrentBalance - WD)
            MM = input("Select 1, 2, 3 or 9")

#Deposits
    while MM =="3":
        print("How much do you want to deposit?")
        DP = int(input("Enter Amount:"))
        if DP > 1000:
            print("Transaction too large, please go to your nearest bank branch")
            MM = input("Select 1, 2, 3 or 9")
            continue
        else:
            print("You now have", CurrentBalance + DP)
            MM = input("Select 1, 2, 3 or 9")

#Exit
    if MM =="9":
        print("Thank you for using The Bank of Joshua")
        return True


Programme(MM)
```
